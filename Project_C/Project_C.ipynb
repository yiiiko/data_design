d.)Plot top 20 words in text

```python
texted = soup.get_text()
words = re.findall('\w+', texted)
lowered = []
for word in words:
    lowered.append(word.lower())
text = nltk.Text(lowered)
fdist = nltk.FreqDist(words)
fdist.plot(20)
```
